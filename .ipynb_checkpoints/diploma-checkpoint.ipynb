```python
def plot_confusion_matrix(y_true, y_pred, classes=[0,1],
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
#     classes = classes[unique_labels(y_true, y_pred)]
#     if normalize:
#         cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
#         print("Normalized confusion matrix")
#     else:
#         print('Confusion matrix, without normalization')

#     print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='Истинное значение',
           xlabel='Предсказанное значение')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

```python
def process_data(df):
    def f1(x):
        if x == 'Infinity':
            return 1
        else:
            return 0
        
        
    def f(x):
        if x == 'BENIGN':
            return 0
        else:
            return 1

    df['Label'] = df['Label'].apply(f)
    
    df['isInfFB'] = df['Flow Bytes/s'].apply(f)

    df['isInfFP'] = df['Flow Packets/s'].apply(f)

    m = pd.to_numeric(df.loc[df['Flow Bytes/s'] != "Infinity", 'Flow Bytes/s']).max()
    df['Flow Bytes/s'].replace("Infinity",m,inplace=True)

    m = pd.to_numeric(df.loc[df['Flow Packets/s'] != "Infinity", 'Flow Packets/s']).max()
    df['Flow Packets/s'].replace("Infinity",m,inplace=True)

    df['Flow Bytes/s'] = pd.to_numeric(df['Flow Bytes/s'])
    df['Flow Packets/s'] = pd.to_numeric(df['Flow Packets/s'])
    
    return df.dropna()
```

```python
model_root_path = './models'
```

```python
def isf_forecast(X_train, X_test, y_test, desc='Model Data Description'):
    
    grid =  {
            "n_estimators" : [25],
            
            "max_features": [0.5],
            "max_samples": [0.8],
            "contamination": [0.3],
            "behaviour": ['new']
    }
    model_zoo = []
    scores = []
    params_zoo = []
    
    for p in tqdm(ParameterGrid(grid)):
        clf = IsolationForest(random_state=42, n_jobs=10, **p)
        clf.fit(X_train)
        y_pred = clf.predict(X_test)
        y_pred = (-y_pred + 1) // 2
        
        model_zoo.append(clf)
        scores.append(f1_score(y_test, y_pred))
        params_zoo.append(p)
        print(p)
        print('f1:',f1_score(y_test, y_pred))
#         print(confusion_matrix(y_test, y_pred))
        
    best = np.argmax(scores)
    model = model_zoo[best]
    params = params_zoo[best]
    
    print(desc)
    y_pred = model.predict(X_test)
    y_pred = (-y_pred + 1) // 2
    print('Best Params:')
    print('n_estimators: ', params)
    print('Test F1 score: {}'.format(round(f1_score(y_pred, y_test), 4))) 
    print('ROC AUC: {}'.format(round(roc_auc_score(y_pred, y_test), 4)))
    print(confusion_matrix(y_test, y_pred))
    
    params_str = ''
    for name, value in zip(params.keys(), params.values()):
        params_str += name + '_' + str(value)
    
    f_name = desc.replace(' ', '_').lower() + '_' + \
    params_str + '_' + 'f1' + '_' + \
    str(round(f1_score(y_pred, y_test), 2)) + '.pkl'
    
    joblib.dump(model, os.path.join(model_root_path, f_name))
    return model
```

```python
anomaly_detector2 = isf_forecast(legal,X_test,y_test,desc='detector')
```

```python
def one_class_svm_forecast(X_train, X_test, y_test, desc='Model Data Description'):
    grid =  {
            'gamma': [0.01,0.05, 0.07, 0.1],
            'nu': [0.001, 0.005],
            'cache_size':[5000],
            'max_iter':[300],
        }
    results = []
    models_zoo = []
    params_zoo = []
    
    for p in tqdm(ParameterGrid(grid)):
        clf = OneClassSVM(**p, random_state=42)
        clf.fit(X_train)
        y_pred = clf.predict(X_test)
        y_pred = (-y_pred + 1) // 2
        results.append(f1_score(y_test, y_pred))
        models_zoo.append(clf)
        params_zoo.append(p)
        print(p)
        print('f1:',f1_score(y_test, y_pred))
        
    
    print(desc)
    best_idx = np.argmax(results)
    
    print('Best Params:')
    print(params_zoo[best_idx])
    print('Best CV F1 Score: {}'.format(round(results[best_idx], 4)))
    
    model = models_zoo[best_idx]
    y_pred = model.predict(X_test)
    y_pred = (-y_pred + 1) // 2
    print('Test F1 score: {}'.format(round(f1_score(y_pred, y_test), 4)))
    print('ROC AUC: {}'.format(round(roc_auc_score(y_pred, y_test), 4)))
    print(confusion_matrix(y_test, y_pred))
    
    f_name = desc.replace(' ', '_').lower() + '_' + \
    'gamma_' + str(params_zoo[best_idx]['gamma']) + '_' + \
    'nu_' + str(params_zoo[best_idx]['nu']) + '_' + \
    'f1_' + str(round(f1_score(y_pred, y_test), 2)) + '.pkl'
    
    joblib.dump(model, os.path.join(model_root_path, f_name))
    return model

def test(path, X_test, y_test):
    model = joblib.load(path)
    out = path.split('/')[-1].replace('_', ' ').upper()
    print(out)
    y_pred = model.predict(X_test)
    y_pred = (-y_pred + 1) // 2
    print('Test F1 score: {}, ROC AUC: {}'.format(round(f1_score(y_pred, y_test), 4), 
                                                  round(roc_auc_score(y_pred, y_test), 4)))
    return model
```
